## Pandas

**Pandas** és la llibreria per excel·lència per a l'anàlisi de dades del llenguatge Python. El seu nom prové de “panell data” (terme economètric). Inspirada en les funcionalitats de fulls de càlcul i _R_, però amb el potencial d'aquest llenguatge de propòsit general.

**Pandas** inclou totes les funcionalitats necessàries per al procés d'anàlisi de dades: càrrega, filtrat, tractament, síntesi, agrupament, emmagatzematge i visualització. A més, s'integra amb la resta de llibreries de càlcul numèric com Numpy, visualització amb Matplotlib, aprenentatge automàtic amb scikit-learn, etc.

**L'objectiu d'aquest curs introductori és adquirir capacitats bàsiques d'anàlisis de dades amb Pandas.**

## Estructura del dataframe

Descobrirem com podem accedir a l'informació d'un dataframe:

### Columnes

Començarem amb les columnes

````python
# Columnes o caracteristiques de cada mostra
print(df_who.columns)
````

````python
for col in df_who.columns:
    print(col)
````

````python
df_who["Country"] # Una serie
````

````python
print(df_who["Country"][0])
print("-"*30)
print(df_who["Country"][:5])
print("-"*30)
print(df_who["Country"].values)
````

````python
df_who.Country # una altra manera més simple
# pero hi ha noms de columnes molt llargues: "Children aged &lt;5 years who received any antimalarial treatment for fever (%)" 
# NOTA: En la creació de documents és important un adequat nom de columnes
````

````python
print(df_who.columns[9])
print("-"*30)
print(df_who[df_who.columns[9]])
````

````python
# Podem agafar multiples columnes
df_who[df_who.columns[0:5]]
````

````python
df_who[["CountryID","Continent"]]
````

````python
# Podem accedir a les columnes amb la seva posició en lloc del nom
# funció: .iloc(files, columnes)
df_who.iloc[:,3:5]
````

````python
df_who.iloc[:,[3,5,9]]
````

### Files

Cada mostra té un índex. L'índex pot ser numèric, alfabètic o de temps.

````python
df_who.index
````

````python
df_who.index.values
````

````python
df_who.loc[0] #la mostra zero
````

````python
type(df_who.loc[0]) # una mostra i una columna són també sèries
````

````python
print(df_who.loc[0].Country) # Podem accedir a una sèrie amb el seu índex (pròximes classes)
print(df_who.loc[0][0]) #o amb la seva posició
print(df_who.loc[0][3])
````

````python
df_who.loc[166:170] #slicing
````

````python
# Els índexs de les mostres són útils quan són dates: Sèries temporals
# per exemple:
# df.loc["2020":"2022"]
````

````python
# Obviament, podem fer una selecció més especifica
df_who.loc[167:169,["Country","Total_CO2_emissions"]] # per noms 
````

````python
df_who.iloc[[168,192],[0,43,118]] # per posicions
````
